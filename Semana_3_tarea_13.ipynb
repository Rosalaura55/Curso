# Spearman correlation

```python
## Spearman Correlation between DIC and Temperature
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import spearmanr
```

```python
df = pd.read_csv("Terminos_lagoon_TA_DIC_2023_RawData.csv")
```

```python
DIC_COL ="dic_micromol_kg"
TEMP_COL = "temp_c"
```

```python
df[DIC_COL]  = pd.to_numeric(df[DIC_COL], errors="coerce")
df[TEMP_COL] = pd.to_numeric(df[TEMP_COL], errors="coerce")
df_clean = df.dropna(subset=[DIC_COL, TEMP_COL])
```

```python
rho, pval = spearmanr(df_clean[DIC_COL], df_clean[TEMP_COL])

print("=== Spearman Correlation between DIC and Temperature ===")
print(f"Spearman rho = {rho:.3f}")
print(f"p-value = {pval:.4f}")

if pval < 0.05:
    print("Result: Significant correlation (p < 0.05)")
else:
    print("Result: No significant correlation (p >= 0.05)")

# Scatterplot with regression line (nonparametric trend) 
plt.figure(figsize=(8,6))
sns.scatterplot(data=df_clean, x=TEMP_COL, y=DIC_COL, alpha=0.7)
sns.regplot(
    data=df_clean, x=TEMP_COL, y=DIC_COL,
    scatter=False, lowess=True, color="red"
)
plt.title("Spearman Correlation: DIC vs Temperature")
plt.xlabel("Temperature (C)")
plt.ylabel("DIC (mol/kg)")
plt.tight_layout()
plt.show()
```

Temperature and dissolved inorganic carbon have a negative correlation (-0.018) that is not significant (p-value = 0.8570 (p > 0.05). Therefore, as temperature increases, DIC decreases in a non-significant manner.
