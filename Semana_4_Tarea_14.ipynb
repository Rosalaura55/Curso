Spearman correlation matrix

```python
import pandas as pd

df = pd.read_csv("Terminos_lagoon_TA_DIC_2023_RawData.csv")
```

```python

chl_dep_dic_df = df[['chlorophy_microg_l', 'depth_m', 'dic_micromol_kg']]
```

```python
corr_matrix = chl_dep_dic_df.corr(method='spearman')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the heatmap 
plt.figure(figsize=(8, 6))
sns.set(style="white", font_scale=1.2)
sns.heatmap(corr_matrix, 
            annot=True, 
            fmt=".2f", 
            cmap='cividis', 
            square=True, 
            linewidths=0.5, 
            linecolor='gray',
            cbar_kws={'shrink': 0.8})

plt.title('Spearman Correlation Matrix', fontsize=14, weight='bold')
plt.tight_layout()
plt.show()
```

There is a positive correlation between depth and chlorophyll, so as depth increases, the amount of chlorophyll also increases. The other variables have a low negative correlation, meaning one increases when the other decreases.
